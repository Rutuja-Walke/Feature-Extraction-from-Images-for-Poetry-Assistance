```python
a_file = open("/content/drive/MyDrive/MajorProject/Dataset/Flickr Datasets/flickr30k_images/30ktoken_1_copy.txt", "r")

list_of_lists = []
for line in a_file:
  stripped_line = line.strip()
  line_list = stripped_line.split('\n')
  list_of_lists.append(line_list)

a_file.close()
```

```python
list_of_lists
```

```python
from collections import Iterable

def flatten(lis):
     for item in lis:
         if isinstance(item, Iterable) and not isinstance(item, str):
             for x in flatten(item):
                 yield x
         else:        
             yield item

```

```python
b = list(flatten(list_of_lists))
```

```python
b
```

```python
b2 = b
```

```python
b3 = []
for i in range(len(b2)):
  #str_temp=""
  #str_temp = b2[i]
  #if(b2[i][0])
  if(b2[i]==""):
    continue
  if b2[i][0] in ['1','2','3','4','5','6','7','8','9']:
    b3.append(b2[i])
```

```python
textfile = open("/content/drive/MyDrive/MajorProject/Dataset/Flickr Datasets/flickr30k_images/30ktoken_2.txt", "w")
for element in b3:
    textfile.write(element + "\n")
textfile.close()
```
